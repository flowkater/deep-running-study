```python
import numpy as np
import pandas as pd
```

# Load Dataset

```python
from keras.datasets import mnist

((X_train, y_train), (X_test, y_test)) = mnist.load_data()
```

# Preprocessing

```python
X_train = X_train.reshape(60000, 28, 28, 1)
X_test = X_test.reshape(10000, 28, 28, 1)

print(X_train.shape, X_test.shape)
```

```python
from keras.utils import to_categorical

y_train_hot = to_categorical(y_train)
y_test_hot = to_categorical(y_test)

print(y_train_hot.shape, y_test_hot.shape)

print("Label = {0}".format(y_train[0:3]))
y_train_hot[0:3]
```

# Build a Model

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten
from keras.optimizers import SGD

# Convolutional Layer == Convolutional Filter == Convolutional Kernel

model = Sequential()

model.add(Conv2D(
    filters=6,
    kernel_size=(5, 5),
    activation='sigmoid',
    kernel_initializer='random_uniform',
    input_shape=(28, 28, 1)
))
model.add(MaxPool2D(
    pool_size=(2, 2)
))
model.add(Conv2D(
    filters=12,
    kernel_size=(5, 5),
    activation='sigmoid',
    kernel_initializer='random_uniform'
))
model.add(MaxPool2D(
    pool_size=(2, 2)
))

model.add(Flatten())
model.add(Dense(
    units=128,
    kernel_initializer='random_uniform',
    activation='sigmoid'
))
model.add(Dense(
    units=10,
    kernel_initializer='random_uniform',
    activation='sigmoid'
))

model.compile(
    loss='categorical_crossentropy',
    optimizer=SGD(lr=0.1),
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train_hot, epochs=20)
```

```python
y_predict = model.predict(X_test)
y_predict = np.argmax(y_predict, axis=1)

accuracy = (y_predict == y_test).mean()

print("Accuracy = {0:.5f}".format(accuracy))
result = pd.DataFrame({'y(actual)': y_test, 'y(predict)': y_predict})
result[result['y(actual)'] != result['y(predict)']]
```
